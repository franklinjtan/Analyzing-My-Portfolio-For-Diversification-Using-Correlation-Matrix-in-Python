#### Import pandas, numpy, seaborn, matplotlib, and scipy libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import seaborn

%matplotlib inline
```

```python
#  Calculate daily percent change
SPY  = pd.read_csv('SPY.csv');  SPY['pct'] = SPY['Adj Close'].pct_change()
AMZN  = pd.read_csv('AMZN.csv');  AMZN['pct'] = AMZN['Adj Close'].pct_change()
CNQ  = pd.read_csv('CNQ.csv');  CNQ['pct'] = CNQ['Adj Close'].pct_change()
FHLC  = pd.read_csv('FHLC.csv');  FHLC['pct'] = FHLC['Adj Close'].pct_change()
FIDU  = pd.read_csv('FIDU.csv');  FIDU['pct'] = FIDU['Adj Close'].pct_change()
FSTA = pd.read_csv('FSTA.csv');  FSTA['pct'] = FSTA['Adj Close'].pct_change()
GLD = pd.read_csv('GLD.csv');  GLD['pct'] = GLD['Adj Close'].pct_change()
GOOG = pd.read_csv('GOOG.csv');  GOOG['pct'] = GOOG['Adj Close'].pct_change()
ONEQ = pd.read_csv('ONEQ.csv');  ONEQ['pct'] = ONEQ['Adj Close'].pct_change()
OXY = pd.read_csv('OXY.csv');  OXY['pct'] = OXY['Adj Close'].pct_change()
SPHQ = pd.read_csv('SPHQ.csv');  SPHQ['pct'] = SPHQ['Adj Close'].pct_change()
VYM = pd.read_csv('VYM.csv');  VYM['pct'] = VYM['Adj Close'].pct_change()
```

```python
#  Extract percent change to numpy vectors and delete first entry (nan)
spy  = SPY['pct'].to_numpy(); spy = np.delete(spy, 0)
amzn  = AMZN['pct'].to_numpy(); amzn = np.delete(amzn, 0)
cnq  = CNQ['pct'].to_numpy(); cnq = np.delete(cnq, 0)
fhlc  = FHLC['pct'].to_numpy(); fhlc = np.delete(fhlc, 0)
fidu  = FIDU['pct'].to_numpy(); fidu = np.delete(fidu, 0)
fsta = FSTA['pct'].to_numpy(); fsta = np.delete(fsta, 0)
gld  = GLD['pct'].to_numpy(); gld = np.delete(gld, 0)
goog  = GOOG['pct'].to_numpy(); goog = np.delete(goog, 0)
oneq = ONEQ['pct'].to_numpy(); oneq = np.delete(oneq, 0)
oxy  = OXY['pct'].to_numpy(); oxy = np.delete(oxy, 0)
sphq  = SPHQ['pct'].to_numpy(); sphq = np.delete(sphq, 0)
vym = VYM['pct'].to_numpy(); vym = np.delete(vym, 0)
```

### AMAZON - CORRELATION WITH SPY

```python
#  Plot raw AMZN Data
plt.plot(spy, amzn, '.')
plt.grid(True)
plt.xlabel('SPY % Change')
plt.ylabel('AMZN % change')

#  Do linear regression calculating the slope (beta) and correlation (stored in r_value)
slope, intercept, r_value, p_value, std_err =  linregress(spy, amzn)

#  Plot regression line
x = np.linspace(-.1, .1)
plt.plot(x, slope * x + intercept, 'k')

#  Print beta and correlation
print('beta = ', slope)
print('Corr = ', r_value)
```

```python
#  Create a data frame made of our percent changes for each underlying
data = [SPY['pct'], AMZN['pct'], CNQ['pct'], FHLC['pct'], FIDU['pct'], FSTA['pct'], GLD['pct'], GOOG['pct'], ONEQ['pct'], OXY['pct'], SPHQ['pct'], VYM['pct']]
headers = ['SPY', 'AMZN', 'CNQ', 'FHLC', 'FIDU', 'FSTA', 'GLD', 'GOOG', 'ONEQ', 'OXY', 'SPHQ', 'VYM']
pct_frame = pd.concat(data, axis=1, keys=headers)

#  Use the built in pandas function to calculate correlation
corr_frame = pct_frame.corr()
corr_frame
```

```python
#  Create a mask to hide upper part of the correlation matrix
mask = np.triu(np.ones_like(corr_frame, dtype=bool))

#  Create heat map
seaborn.heatmap(corr_frame, cmap='YlGnBu', mask = mask)
```
